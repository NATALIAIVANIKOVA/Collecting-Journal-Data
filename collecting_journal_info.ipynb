```python
from typing import List
```

```python
def getJournalPageLinks(page_number: int) -> List[str]:
    page_link = f'https://cyberleninka.ru/journal/{page_number}'
    response = requests.get(page_link, headers={'User-Agent': UserAgent().safari})
    if not response.ok:
        return []
    html = response.content
    soup = BeautifulSoup(html,'html.parser')
    hrefs = [a['href'] for a in soup.find_all('a') if 'href' in a.attrs]
    return hrefs
```

```python
journal_links_page_2 = getJournalPageLinks(2)
print(len(journal_links_page_2))
```

```python
journal_links_page_2
```

```python
def add_page_prefix(prefix: str, list_of_urls: List[str]) -> List[str]:
    extended_urls = []
    for url in list_of_urls:
        extended_urls.append(prefix + url)
    return extended_urls
```

```python
journal_prefix = 'https://cyberleninka.ru'
journal_links_page_2_extended = add_page_prefix(prefix=journal_prefix, list_of_urls=journal_links_page_2)
```

```python
journal_links = []
journal_prefix = 'https://cyberleninka.ru'
page_number = 1
num_subsequent_errors = 0
while True:
    if page_number % 5 == 0:
        print(f'processing page {page_number} ...')
    journal_links_from_current_page = getJournalPageLinks(page_number=page_number)
    if journal_links_from_current_page != []:
        journal_links_from_current_page = add_page_prefix(prefix=journal_prefix, list_of_urls=journal_links_from_current_page)
        journal_links.extend(journal_links_from_current_page)
        num_subsequent_errors = 0
        page_number += 1
    else:
        print(f'Page {page_number} could not be read or it does not exist!')
        num_subsequent_errors += 1
        page_number += 1
        if num_subsequent_errors == 3:
            print("Stopping due to 3 consecutive errors.")
            break
```

```python
def getStats(soup) -> dict:
    journal_main = soup.find('div', attrs = {'class' : 'main'})
    name = journal_main.find('h1').contents[0]
    description = journal_main.find('p').text.strip()
    if not description:
      description = 'Без описания'
    journal_ranking = journal_main.find('div', attrs = {'class' :'labels'})
    scopus = journal_ranking.find('div', attrs={'class' : 'label scopus'})
    if scopus:
      in_scopus = scopus.text.strip()
    if not scopus:
      in_scopus = 'Не в Scopus'
    vak = journal_ranking.find('div', attrs={'class' : 'label vak'})
    if vak:
      in_vak = vak.text.strip()
    if not vak:
      in_vak = 'Не в ВАК'
    rsci = journal_ranking.find('div', attrs={'class' : 'label rsci'})
    if rsci:
      in_rsci = rsci.text.strip()
    if not rsci:
      in_rsci = 'Не в RSCI'
    stats = journal_main.find('div', attrs={'class' : 'statistics'})
    views = stats.find('div', attrs = {'title' : 'Просмотрели статей'}).find('span').text.strip()
    Hirsch = stats.find('div', attrs = {'title' : 'Индекс Хирша'}).find('span').text.strip()
    return {
        'Наименование издания': name,
        'Описание': description,
        'Индексация в Scopus': in_scopus,
        'Наличие в списке ВАК': in_vak,
        'Индексация в RSCI': in_rsci,
        'Количество просмотров статей': views,
        'Индекс Хирша' : Hirsch
    }
```

```python
result_journal_1 = getStats(soup_1)
result_journal_1
```

```python
def getJournalData(journal_link: str) -> dict:
    response = requests.get(journal_link, headers={'User-Agent': UserAgent().safari})
    if not response.ok:
        return response.status_code
    html = response.content
    soup = BeautifulSoup(html,'html.parser')
    journal_row = getStats(soup)
    journal_row['Ссылка на страницу журнала'] = link
    return journal_row
```

```python
from tqdm import tqdm, tqdm_notebook
```

```python
import time
```

```python
data = []
counter = 1
for link in tqdm_notebook(journal_links):
    try:
        journal_row = getJournalData(link)
        data.append(journal_row)
        time.sleep(0.5)
        if counter % 100 == 0:
            print(f"OK after {counter} journal pages")
        counter += 1
    except Exception as e:
        print(f"Exception occurred with {link}: {e}")
        counter += 1
        continue
```

```python
journal_info = pd.DataFrame(data)
```
